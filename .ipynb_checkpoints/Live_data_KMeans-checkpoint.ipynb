```python
df= data
df.head()
```

```python
# Assign the variables for the independent and dependent variables
X = df
y = df['status_type']
```

```python
print(X.head())
```

```python
print(y.head())
```

```python
# Encode the categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['status_type']= le.fit_transform(X['status_type'])
y = le.transform(y)
```

```python
X.head()
```

```python
y
```

```python
X.info()
```

```python
cols = X.columns
```

```python
# Feature Scaling (Standardizing the dataset)
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
X = mms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
X.head()
```

### KMeans Clustering & Elbow Method

```python
# KMeans Clustering

# Training the Model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
labels = kmeans.labels_
correct_labels = sum(y == labels)
print("Result: %d out of %d samples are correctly labeled."%
     (correct_labels,y.size))
```

```python
print('Accuracy Score: {0:0.2f}'.format(correct_labels/float(y.size)))
```

```python
# Find optimal clusters using Elbow method

from sklearn.cluster import KMeans
cs=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1,11),cs)
plt.title('The Elbow point')
plt.xlabel('Number of clusters')
plt.ylabel('cs')
plt.show()
```

### Final KMeans Fit, Label Alignment & Evaluation

```python
kmeans = KMeans(n_clusters=3,random_state=0)
kmeans.fit(X)
labels=kmeans.labels_
correct_labels=sum(y==labels)
print("Result: %d out of %d are labeled correctly."%
      (correct_labels,y.size))
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
kmeans.fit(X)
labels=kmeans.labels_
correct_labels=sum(y==labels)
print("Result: %d out of %d are correctly labeled."%
     (correct_labels,y.size))
print('Accuracy score: {0:0.2f}'.format(correct_labels/float(y.size)))
```

```python
from scipy.stats import mode

# Fit K-Means
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_

# Relabel clusters to match true labels
new_labels = np.zeros_like(labels)
for i in range(4):  # Assuming 4 clusters
    mask = (labels == i)
    new_labels[mask] = mode(y[mask])[0]

# Compute accuracy
correct_labels = sum(y == new_labels)
accuracy = correct_labels / y.size

print(f"Result: {correct_labels} out of {y.size} are correctly labeled.")
print(f'Accuracy score: {accuracy:.2f}')
```

# Conclusion

#Model Accuracy & Labeling Performance
94% Accuracy Score (6622 out of 7050 correctly labeled):
The model performed exceptionally well when k=4, achieving a high clustering accuracy of 0.94 (94%).
This indicates that the dataset is naturally suited for four distinct groups.

#Optimal Number of Clusters (Elbow Method)
The Elbow Point graph initially showed a steep decline in inertia up to k=2, but performance improved significantly at k=4.
k=4 was the optimal choice, leading to well-defined and meaningful clusters.
Choosing fewer clusters (e.g., k=2) resulted in poor accuracy and weak cluster separation.

#Model Performance Insights
Poor performance with k=2 suggests that the dataset had more than two meaningful groups.
k=4 provided the best structure, meaning the model successfully identified four well-separated groups.
The high accuracy confirms that the dataset features support this cluster distribution.
