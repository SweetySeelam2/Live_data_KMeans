# üìä Live Social Media Post Segmentation with K-Means Clustering
### Uncovering Patterns in User Engagement to Optimize Content Strategy

---

## üè¢ Business Problem

Modern brands invest heavily in social media, but most struggle to identify *which types of posts drive meaningful engagement* and why. Raw metrics alone (likes, shares, reactions) don‚Äôt reveal *actionable segments* among your content types.

*How can we automatically segment posts to uncover engagement patterns and optimize content for maximum reach and brand growth?*

## üéØ Objective

- *Cluster social media posts* based on engagement metrics (likes, shares, reactions, etc.) to discover high-engagement segments.
- Provide actionable, data-driven recommendations for content strategy.
- Demonstrate unsupervised learning (K-Means) on real engagement data.

- **Source**: [UCI Machine Learning Repository ‚Äî Facebook Live Posts Dataset](https://archive.ics.uci.edu/ml/datasets/Facebook+Live+Sellers+in+Thailand)

## üßπ Step 3: Data Cleaning & Preprocessing

- Remove unused columns (`Column1-4`, `status_id`, `status_published`)
- Encode categorical variables
- Feature scale for KMeans

```python
# Drop unused columns (all-NaN columns, ID and timestamp columns)
data = data.drop(['Column1','Column2','Column3','Column4','status_id','status_published'], axis=1)

# Encode categorical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['status_type'] = le.fit_transform(data['status_type'])
status_type_labels = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label encoding mapping:", status_type_labels)
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(data)
X = pd.DataFrame(X, columns=data.columns)
```

----

## üîé Step 4: KMeans Clustering & Elbow Method

```python
# 1. Find optimal clusters using Elbow method

from sklearn.cluster import KMeans

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(6,5))
plt.plot(range(1, 11), inertias, marker='o')
plt.title('The Elbow Point')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# 2. Fit KMeans on Only Feature Columns Without Any Labels

from sklearn.cluster import KMeans

# Choose number of clusters (elbow, silhouette, business intuition, etc.)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(X)
```

```python
# 3. Profile Each Cluster

# After fitting kmeans on X (WITHOUT 'cluster')
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)
display(cluster_centers)

# Add cluster labels to a copy for analysis
X_unlabeled = X.copy()
X_unlabeled['cluster'] = cluster_labels
```

```python
# Cluster sizes
print("Cluster sizes:")
print(X_unlabeled['cluster'].value_counts())
```

```python
# Feature means per cluster
cluster_summary = X_unlabeled.groupby('cluster').mean()
display(cluster_summary)
```

```python
# 4. Visualize/Interpret Each Cluster
```

```python
# 4A. Feature Means per Cluster

import matplotlib.pyplot as plt
import seaborn as sns

cluster_summary = X_unlabeled.groupby('cluster').mean()

plt.figure(figsize=(12, 6))
sns.heatmap(cluster_summary.T, annot=True, cmap="YlOrRd")
plt.title("Feature Means per Cluster (Unsupervised KMeans)")
plt.ylabel("Features")
plt.xlabel("Cluster")
plt.show()
```

```python
# 4B. Boxplots or Pairplots by Cluster

# Boxplots
for col in ['num_reactions', 'num_comments', 'num_shares', 'num_likes']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='cluster', y=col, data=X_unlabeled)
    plt.title(f'{col} Distribution by Cluster')
    plt.show()
```

```python
# 4C. Cluster Sizes (Bar Chart)

X_unlabeled['cluster'].value_counts().plot(kind='bar', color='dodgerblue')
plt.title('Number of Posts per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()
```

----

## ü§ñ Step 5: Final Model: Clustering & Label Alignment

```python
# Considering 2 clusters
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(X)

from scipy.stats import mode

# Relabel clusters to best match the true status_type labels for accuracy assessment
y_true = data['status_type'].values
new_labels = np.zeros_like(labels)
for i in range(n_clusters):
    mask = (labels == i)
    if np.sum(mask) == 0:
        continue
    new_labels[mask] = mode(y_true[mask], keepdims=True)[0]

# Accuracy score (not perfect for unsupervised but a useful indicator)
accuracy = np.sum(y_true == new_labels) / len(y_true)
print(f"Clustering accuracy (after relabeling): {accuracy:.2f}")
```

```python
# Considering 3 clusters
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(X)

from scipy.stats import mode

# Relabel clusters to best match the true status_type labels for accuracy assessment
y_true = data['status_type'].values
new_labels = np.zeros_like(labels)
for i in range(n_clusters):
    mask = (labels == i)
    if np.sum(mask) == 0:
        continue
    new_labels[mask] = mode(y_true[mask], keepdims=True)[0]

# Accuracy score (not perfect for unsupervised but a useful indicator)
accuracy = np.sum(y_true == new_labels) / len(y_true)
print(f"Clustering accuracy (after relabeling): {accuracy:.2f}")
```

```python
# Considering 4 clusters; 4 is chosen since we have 4 post types
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(X)

from scipy.stats import mode

# Relabel clusters to best match the true status_type labels for accuracy assessment
y_true = data['status_type'].values
new_labels = np.zeros_like(labels)
for i in range(n_clusters):
    mask = (labels == i)
    if np.sum(mask) == 0:
        continue
    new_labels[mask] = mode(y_true[mask], keepdims=True)[0]

# Accuracy score (not perfect for unsupervised but a useful indicator)
accuracy = np.sum(y_true == new_labels) / len(y_true)
print(f"Clustering accuracy (after relabeling): {accuracy:.2f}")
```

-----------

## üìä Step 6: Cluster Analysis & Interpretation

```python
# Append cluster labels for EDA
X['cluster'] = new_labels

# Compare mean engagement metrics across clusters
cluster_means = X.groupby('cluster').mean()
cluster_means[['num_reactions','num_comments','num_shares','num_likes']].plot(kind='bar', figsize=(10,6))
plt.title('Average Engagement Metrics per Cluster')
plt.ylabel('Mean Value (Scaled)')
plt.xticks(rotation=0)
plt.show()
```

```python
# Heatmap of cluster centers
plt.figure(figsize=(8,4))
sns.heatmap(cluster_means.T, annot=True, fmt=".2f", cmap="YlOrRd")
plt.title("Feature Means per Cluster")
plt.ylabel("Features")
plt.xlabel("Cluster")
plt.show()
```
