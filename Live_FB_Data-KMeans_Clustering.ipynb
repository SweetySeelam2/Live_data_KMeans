## üßπ Step 3: Data Cleaning & Preprocessing

```python
# Drop unused columns (all-NaN columns, ID and timestamp columns)
data = data.drop(['Column1','Column2','Column3','Column4','status_id','status_published'], axis=1)

# Encode categorical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['status_type'] = le.fit_transform(data['status_type'])
status_type_labels = dict(zip(le.classes_, le.transform(le.classes_)))
print("Label encoding mapping:", status_type_labels)
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(data)
X = pd.DataFrame(X, columns=data.columns)
```

## üîé Step 4: KMeans Clustering & Elbow Method

```python
# 1. Find optimal clusters using Elbow method

from sklearn.cluster import KMeans

inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(6,5))
plt.plot(range(1, 11), inertias, marker='o')
plt.title('The Elbow Point')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

**Results Interpretation - Elbow Point Plot**

Line plot: Inertia vs. Number of clusters

- The elbow in the plot occurs around k=3 to 4, after which the decrease in inertia slows, suggesting 3‚Äì4 clusters is optimal for your data.

- Beyond 4 clusters, additional clusters don‚Äôt significantly improve model fit.

*Business Insight:*
Choosing 3 or 4 clusters balances model simplicity with capturing key engagement patterns, making insights actionable and not over-complicated.

```python
# 2. Fit KMeans on Only Feature Columns Without Any Labels

from sklearn.cluster import KMeans

# Choose number of clusters (elbow, silhouette, business intuition, etc.)
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
cluster_labels = kmeans.fit_predict(X)
```

```python
# 3. Profile Each Cluster

# After fitting kmeans on X (WITHOUT 'cluster')
cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)
display(cluster_centers)

# Add cluster labels to a copy for analysis
X_unlabeled = X.copy()
X_unlabeled['cluster'] = cluster_labels
```

## ü§ñ Step 5: Final Model: Clustering & Label Alignment

```python
# Considering 2 clusters
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(X)

from scipy.stats import mode

# Relabel clusters to best match the true status_type labels for accuracy assessment
y_true = data['status_type'].values
new_labels = np.zeros_like(labels)
for i in range(n_clusters):
    mask = (labels == i)
    if np.sum(mask) == 0:
        continue
    new_labels[mask] = mode(y_true[mask], keepdims=True)[0]

# Accuracy score (not perfect for unsupervised but a useful indicator)
accuracy = np.sum(y_true == new_labels) / len(y_true)
print(f"Clustering accuracy (after relabeling): {accuracy:.2f}")
```

## üöÄ Step 7: Random Forest Classifier + SHAP & LIME Explainability

**1. Why Add a Classifier to Unsupervised KMeans?**

**Why Are We Using a Classifier Here?**

KMeans is an unsupervised learning algorithm:  
- It segments data into clusters purely by patterns in the features‚Äî**without using any human-provided labels or target variables**.
- However, KMeans alone doesn't tell us *which* features are most important for each cluster, or *why* a post was assigned to a particular group.

**Goal:**  
To make the clusters actionable and trustworthy for business users, we want to:
- Understand the *feature drivers* for each cluster.
- Provide transparent, human-interpretable explanations for each assignment.

**How?**  
We train a supervised classifier (e.g., RandomForest) to "mimic" KMeans‚Äîthat is, predict the cluster assignment from the original feature set.  
- Once trained, we can use state-of-the-art explainability tools like **SHAP** and **LIME** to see which features are most influential for each cluster.

**Result:**  
- Marketers, analysts, and business users gain insight into *what defines each cluster*.
- You bridge the gap between "black box" AI and real-world, transparent business strategy.

---

**2. Purpose & Value of SHAP/LIME in Clustering**

**Why SHAP & LIME for This Project?**

- SHAP (SHapley Additive exPlanations) and LIME (Local Interpretable Model-agnostic Explanations) are designed for *supervised* models.
- By training a classifier on the K-Means assignments, we can use SHAP/LIME to:
    - Identify the *most important features* for each cluster (global interpretability).
    - Explain **why* a specific post was assigned to a cluster (local interpretability).
- This creates a high-value, enterprise-ready workflow where every cluster is not only detected, but its *business logic* is transparent and actionable.

**Objective:**  
- To provide *trustworthy* and *actionable* AI insights for content strategy and marketing investment.

**Business Impact:**  
- More informed campaign planning.
- Easier buy-in from stakeholders (managers, clients, regulators).
- Model adoption at scale with confidence.

**3. Applying Random Forest classifier model**

```python
# 3.1. Preprocess
feature_cols = [
    'status_type', 'num_reactions', 'num_comments', 'num_shares',
    'num_likes', 'num_loves', 'num_wows', 'num_hahas', 'num_sads', 'num_angrys'
]
X = data[feature_cols]
```

```python
# 3.2. Clustering (Unsupervised)
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
cluster_labels = kmeans.fit_predict(X)
```

```python
# 3.3. Train/Test Split (ONLY ONCE)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, cluster_labels, test_size=0.2, random_state=42, stratify=cluster_labels
)
```

```python
# 3.4. RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=0, class_weight='balanced')
rf.fit(X_train, y_train)
print("Classifier accuracy:", rf.score(X_test, y_test))
```

```python
import joblib

# Save the trained RandomForestClassifier to a file
joblib.dump(rf, "rf_classifier.pkl")
```

```python
# 3.5. SHAP Explainability
import shap
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(X_test)
print("shap_values.shape:", shap_values.shape)

for i in range(4):
    print(f"SHAP summary for cluster {i}")
    shap.summary_plot(shap_values[:, :, i], X_test, feature_names=feature_cols)
```

```python
# 3.6. LIME Explainability (no warning)
from lime.lime_tabular import LimeTabularExplainer
explainer_lime = LimeTabularExplainer(
    X_train.values,
    feature_names=feature_cols,
    class_names=[f'Cluster {i}' for i in range(4)],
    discretize_continuous=True
)
idx = 0
exp = explainer_lime.explain_instance(X_test.iloc[idx].values, rf.predict_proba, num_features=5)
exp.show_in_notebook(show_table=True)
```

- The predicted probability for Cluster 0 is 1.00, making it a highly confident assignment.

- This local explanation matches the global SHAP findings, showing that high comment and share counts strongly determine cluster membership for this post.

*Business Insight:*
LIME explains why a specific post was grouped as it was, confirming that massive engagement in terms of comments and shares drives cluster assignment. Tailoring content to boost these features can help ensure more posts enter high-performing clusters.
